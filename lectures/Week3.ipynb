> * What are the three slope dependent regimes of complex networks with power-law degree distributions? Briefly describe each one. (You will have to skim chp 4.7 to answer this one).\

**Answer**  Anomalous Regime (γ≤ 2)\
 The exponent is larger than one, hence the number of links connected to the largest hub grows faster than the size of the network. This means that for sufficiently large _ N_ the degree of the largest hub must exceed the total number of nodes in the network, hence it will run out of nodes to connect to.\
Scale-Free Regime (2 ‹ γ ‹ 3)\\

scalefree networks in this regime are ultra-small.\\
Random Network Regime (γ › 3)\
 the properties of a scale-free network in this regime are difficult to distinguish from the properties a random network of similar size.\\

> * What are the three regimes we find in non-linear preferential attachement? (chapter 5) Briefly describe each one. that _kmax_ grows with the size of the network with exponent 1/(_γ_ - 1), which is smaller than one. Hence the market share of the largest hub, _kmax_ /_N_, representing the fraction of nodes that connect to it,\\

**Answer** Sublinear Preferential Attachment (0 < α < 1)\
For any α > 0 new nodes favor the more connected nodes over the less connected nodes. Yet, for α < 1 the bias is weak, not sufficient to generate a scale-free degree distribution.\\
Superlinear Preferential Attachment (α > 1)\\
For α > 1 the tendency to link to highly connected nodes is enhanced, accelerating the rich-gets-richer process. The consequence of this is most obvious for α > 2, when the model predicts a winner-takes-all phenomenon: almost all nodes connect to a few super-hubs.\\



> We're going to create our own Barabasi-Albert model (a special case) in right in a `notebook`. Follow the recipe below for success:
> 
> * First create a graph consisting of a single link. (You can call the nodes anything, but I would simply use integers as names).
> * Now add another node, connecting one of the existing nodes in proportion to their degree.
> * Keep going until you have a 100 node network.
>   * *Hint*: The difficult part here is connecting to each node according to their degree. The way I do it is: generate a list of all edges (e.g. pairs of nodes), then flatten it (e.g. remove connection information). That list contains each node in proportion to its connections, thus drawing a random node from that list (e.g. using `random.choice`) corresponds to selecting a node with probability proportional to it's degree.
> * Plot the network. 
> * Add more nodes until you have a 5000 node network.
> * What's the maximum and minimum degree?
> * Now, bin the degree distribution using `numpy.histogram`.

```python
np.random.seed(42)
'''First create a graph consisting of a single link. 
(You can call the nodes anything, but I would simply use integers as names).'''
G = nx.Graph()
G.add_node(1)
G.add_node(2)
G.add_edge(1,2)
'''Now add another node, connecting one of the existing nodes in proportion to 
their degree.Now add another node, connecting one of the existing nodes in proportion to their degree.'''
degree_sum = sum([G.degree(x) for x in G.nodes()])
def Preferential_attachment(node,degree_sum):
    return G.degree(node)/degree_sum

def add_node_ba(G,nodeToAdd,degree_sum=degree_sum):
    nodes_in_graph = list(G.nodes())
    probability = [Preferential_attachment(x, degree_sum) for x in nodes_in_graph]
    nodes_to_link = np.random.choice(nodes_in_graph,size=2,replace=False,p=probability)
    for node in nodes_to_link:
        G.add_edge(nodeToAdd,node)
    return degree_sum+4
    
degree_sum = add_node_ba(G,3)
'''Keep going until you have a 100 node network.'''
last_added_node =3
while(len(list(G.nodes()))<=100):
    degree_sum = add_node_ba(G,last_added_node+1,degree_sum=degree_sum)
    last_added_node+=1

''' Plot the network'''
plt.figure(figsize=(12, 30))
plt1= plt.subplot(5, 1, 1)
nx.draw(G, with_labels=False, width=0.5, nodelist=dict(G.degree),node_size=[dict(G.degree)[k]*3 for k in dict(G.degree)])
plt.title("Barabasi-Albert Network with 100 nodes")

'''Add more nodes until you have a 5000 node network.
'''
while(len(list(G.nodes()))<=5000):
    degree_sum = add_node_ba(G,last_added_node+1,degree_sum=degree_sum)
    last_added_node+=1


plt2= plt.subplot(5, 1, 2)
nx.draw_circular(G, with_labels=False, width=0.1, nodelist=dict(G.degree),node_size=[dict(G.degree)[k]*1.1 for k in dict(G.degree)])
plt.title("Barabasi-Albert Network with 5000 nodes")


'''What's the maximum and minimum degree'''
maxD = max(dict(G.degree).values())
minD= min(dict(G.degree).values())
print("The maximum degree", maxD)
print("The maximum degree", minD)
'''Now, bin the degree distribution using `numpy.histogram'''
res = np.histogram(list(dict(G.degree).values()),bins=20)


plt3= plt.subplot(5, 1, 3)
plt.hist(list(dict(G.degree).values()),bins=res[1],edgecolor='black')
plt.title("Degree Distribution Histogram")
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.7)


'''Plot the distribution. Plot it with both linear and log-log axes.'''
m=2  #The coefficient m describes that each new node arrives with m links.
k=np.arange(minD,maxD)
linearP = [(2*m*(m+1))/(ki*(ki+1)*(ki+2)) for ki in k] #(5.11)

logk= np.log10(k)
logP = np.log10(np.array(linearP))

plt4= plt.subplot(5, 1, 4)
plt.plot(k,linearP)
plt.title("Linear Plot of Degree Distribution")
plt.xlabel("Degree (k)")
plt.ylabel("P(k)")
plt.grid(True, linestyle='--', alpha=0.7)


plt5= plt.subplot(5, 1, 5)
plt.plot(logk,logP)
plt.title("Log-Log Plot of Degree Distribution")
plt.xlabel("log(Degree (k))")
plt.ylabel("log(P(k))")
plt.grid(True, linestyle='--', alpha=0.7)
```

# Part 3: The friendship paradox in scale-free networks

Next step is to explore the [Friendship paradox](https://en.wikipedia.org/wiki/Friendship_paradox). This paradox states that _almost everyone_ has fewer friends than their friends have, on average\*. This sounds crazy, but is actually an almost trivial consequence of living in a social network with a power-law degree distribution. The explanation is that almost everyone is friends with a hub, that drives up the average degree of the friends.
